```python
y.value_counts()
```

```python
def isqrt(n):
    x = n
    y = (x + 1) // 2
    while y < x:
        x = y
        y = (x + n // x) // 2
    return x
```

```python
k = isqrt(len(X))
```

```python
knn = KNeighborsClassifier(n_neighbors=k)
```

```python
knn.fit(X_train, y_train)
```
